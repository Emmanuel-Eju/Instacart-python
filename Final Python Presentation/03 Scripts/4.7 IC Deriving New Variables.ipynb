# If-Statement with User-Defined Functions

```python
#create a subset
df = orders_products_merged[:1000000]
```

```python
df.shape
```

```python
#define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply a function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# Question 4

```python
# checking the frequency of orders by hour 

orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# grouping hours into time periods of the day

hour = []

for x in orders_products_merged['order_hour_of_day']:
  if x in range (0,6):
    hour.append("12am-5am")
  elif x in range (6,12):
    hour.append("6am-11am")
  elif x in range (12,18):
    hour.append("12pm-5pm")
  else:
    hour.append("6pm-11pm")
```

```python
# merging the time periods to orders_products_merged 

orders_products_merged ['period_of_day'] = hour
```

```python
orders_products_merged.head(2)
```

```python
# Busiest period of day is 12pm - 5pm. Slowest period of day is 12am-5am. 

orders_products_merged['period_of_day'].value_counts(dropna = False)
```

```python
#  busiest periods of the day 

busiest_periods = []

for x in orders_products_merged['order_hour_of_day']:
  if x in range (12,18):
    busiest_periods.append("Most orders")
  elif x in range (0,6):
    busiest_periods.append("Fewest orders")
  else:
    busiest_periods.append("Average orders")
```

```python
# merging busiest_periods to orders_products_merged

orders_products_merged ['busiest_periods'] = busiest_periods
```

```python
# checking values counts of busiest_period column

orders_products_merged['busiest_periods'].value_counts(dropna = False)
```
